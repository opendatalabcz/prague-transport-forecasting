```python
import folium
import csv
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
m = folium.Map(location=[50.08, 14.42], zoom_start=13)

with open('cyclo-counters.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        lat = float(row['latitude']) if row['latitude'] else None
        lon = float(row['longitude']) if row['longitude'] else None
        if lat and lon:
            popup = f"<b>{row['name']}</b><br>Trasa {row['route']}"
            folium.Marker(location=[lat, lon], popup=popup).add_to(m)

m.save('mapa_scitacu.html')
```

```python
df = pd.read_csv('detections.csv')
display(df)
display(df.describe())
display(df.info())
```

```python
empty = [col for col in df.columns if df[col].isnull().all()]
df.drop(empty, axis=1, inplace=True)
```

```python
df['date'] = pd.to_datetime(df.date)
```

```python
df.info()
```

```python
invalid_counts = ((df.isna()) | (df == 0)).sum()
invalid_ratio = invalid_counts / len(df)

valid_columns = invalid_ratio[invalid_ratio <= 0.5].index
df = df[valid_columns]

display(df.info())
```

```python
df['sum'] = df.iloc[:, 1:].sum(axis=1)
```

```python
df.to_csv('counts_cleaned.csv')
```
